```python
'''Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?
 Answer:
  In hypothesis testing, errors occur when incorrect conclusions are drawn:

Type I Error (False Positive): Rejecting a true null hypothesis.
Example: Concluding a medicine works when it actually does not.
Probability = α (significance level).
Type II Error (False Negative): Failing to reject a false null hypothesis.
Example: Concluding a medicine does not work when it actually does.
Probability = β.

Impact on decision-making:

Type I errors may lead to unnecessary actions or false claims.
Type II errors may result in missed opportunities or failure to detect real effects.
Balancing both is crucial depending on the situation.
'''
```

```python
'''Question 2: What is the P-value in hypothesis testing, and how should it be interpreted?

Answer:
The P-value is the probability of obtaining results at least as extreme as the observed results, assuming the null hypothesis is true.

Interpretation:

If P-value ≤ α (e.g., 0.05): Reject the null hypothesis.
If P-value > α: Fail to reject the null hypothesis.

It does not prove the null hypothesis is true or false; it only indicates the strength of evidence against it.'''
```

```python
'''Question 3: Difference between Z-test and T-test

Answer:

Feature	=                  Z-test	                     T-test
Sample Size	=              Large (n ≥ 30)	             Small (n < 30)
Population Variance	 =     Known	                       Unknown
Distribution	=            Normal	                     Student’s t-distribution

When to use:

Use Z-test when variance is known and sample size is large.
Use T-test when variance is unknown and sample size is small.'''
```

```python
'''Question 4: Confidence Interval and Margin of Error

Answer:
A confidence interval (CI) is a range of values used to estimate a population parameter.

Formula:
CI = Sample Mean ± Margin of Error

Margin of Error (ME):

Depends on standard deviation, sample size, and confidence level.
Larger ME → wider interval
Smaller ME → narrower interval

Interpretation:
A 95% CI means we are 95% confident that the true population parameter lies within the interval.'''
```

```python
'''Question 6: Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.
Answer:'''

import numpy as np
from scipy import stats

# Sample data
data = [50, 52, 49, 51, 53, 48, 50]
population_mean = 50

# Calculate mean and standard deviation
sample_mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)

# Z-test calculation
z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

#Interpretation:
#If P-value < 0.05, the sample mean significantly differs from the population mean.
```

```python
'''Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.'''
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5

# Generate data
data = np.random.binomial(n, p, 1000)

# Plot histogram
plt.hist(data, bins=11, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
'''Question 8: Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python. '''
import numpy as np
import matplotlib.pyplot as plt

# Generate non-normal data (exponential)
data = np.random.exponential(scale=2, size=10000)

sample_means = []

# Generate samples
for _ in range(1000):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))

# Plot
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()
```

```python
'''Question 9: Write a Python function to calculate and visualize the confidence interval
for a sample mean. '''
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2, len(data)-1)

    lower = mean - margin
    upper = mean + margin

    print("Confidence Interval:", (lower, upper))

    # Plot
    plt.errorbar(1, mean, yerr=margin, fmt='o')
    plt.title("Confidence Interval")
    plt.xlim(0, 2)
    plt.show()

# Example data
data = np.random.normal(50, 5, 30)
confidence_interval(data)
```

```python
'''Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.'''

from scipy.stats import chisquare

# Observed and expected frequencies
observed = [20, 30, 25, 25]
expected = [25, 25, 25, 25]

# Chi-square test
chi_stat, p_value = chisquare(observed, expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject null hypothesis: distributions differ")
else:
    print("Fail to reject null hypothesis: distributions are similar")

#Explanation:

#If P-value < 0.05 → observed distribution differs from expected
#Otherwise → no significant difference
```
